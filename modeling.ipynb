```python
#Import all required libraries
import pandas as pd
import numpy as np
np.random.seed(0)
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LassoCV, RidgeCV
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from math import sqrt
```

# Read train and test summary dataset

```python
df1 = pd.read_csv("summary_train_portfolios.csv", index_col='ID')
df1
```

```python
df2 = pd.read_csv("summary_test_portfolios.csv",index_col='ID')
df2
```

```python
# Create numpy arrays with  the input features and the target. assign to X and Y
X_train = df1.drop(['ln_LR'],axis=1).values
Y_train = df1['ln_LR'].values
```

# Standardize the data

```python
# standardize the input attributes
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)

print('mean(X_train_scaled)=', X_train_scaled.mean(axis=0))
print('std(X_train_scaled)=', X_train_scaled.std(axis=0))
```

```python
# Create numpy arrays with  the test features and the target. assign to X and Y
X_test = df2.values
```

```python
# standardize the testing attributes
scaler = StandardScaler()
scaler.fit(X_test)
X_test_scaled = scaler.transform(X_test)
```

# Modeling

```python
# Ridge with 10 fold cross validation
l2_CV = RidgeCV(cv=10,alphas=np.linspace(0.001,0.1,100))
l2_CV.fit(X_train_scaled, Y_train)
print('alpha = ', l2_CV.alpha_) 
print('coef = ', l2_CV.coef_)
```

```python
# Ridge (best alpha) - training RMSE
y_pred_train_l2 = l2_CV.predict(X_train_scaled)
train_MSE_l2 = mean_squared_error(Y_train, y_pred_train_l2)
train_RMSE_l2 = sqrt(train_MSE_l2)
print('\ntraining MSE obtained in training dataset Ridge Regression (best alpha) = ', train_MSE_l2)
# print('training RMSE obtained in training dataset Ridge Regression (best alpha) = ', train_RMSE_l2)


```

```python
# Finally, predict the ln(Loss Ratio) for test dataset  using the built model

# Ridge (best alpha) - predict
y_pred_test_l2 = l2_CV.predict(X_test_scaled)
```

```python
df2['ln_LR']= y_pred_test_l2
print(df2[['ln_LR']])
```

```python
df2[['ln_LR']].to_csv("predictions.csv", index=True)
print("Done!")
```
