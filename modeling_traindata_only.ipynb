```python
#Import all required libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LassoCV, RidgeCV
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from math import sqrt
```

# Read train and test summary dataset

```python
df1 = pd.read_csv("summary_train_portfolios.csv", index_col='ID')
df1
```

```python
# Create numpy arrays with  the input features and the target. assign to X and Y
X = df1.drop(['ln_LR'],axis=1).values
Y = df1['ln_LR'].values
```

# Standardize the data

```python
# standardize the input attributes
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)

print('mean(X_scaled)=', X_scaled.mean(axis=0))
print('std(X_scaled)=', X_scaled.std(axis=0))
```

```python
#  Split the data into training and testing.
X_train_scaled, X_test_scaled, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.33, shuffle= False)
```

# Modeling

```python
# Apply linear model with intercept
lr = linear_model.LinearRegression(fit_intercept=True)
```

```python
# Using the training dataset fit LinearRegression model
lr.fit(X_train_scaled, Y_train)
```

```python
# print intercept and the coefficients associated with the input attributes
print('w0 = ', lr.intercept_)
print('w1 = ', lr.coef_)
```

```python
# Lasso with 10 fold cross validation
l1_CV=LassoCV(cv=10,alphas=np.linspace(0.001,0.1,100))
l1_CV.fit(X_train_scaled, Y_train)
print('alpha = ', l1_CV.alpha_)
print('coef = ', l1_CV.coef_)
```

```python
# Ridge with 10 fold cross validation
l2_CV = RidgeCV(cv=10,alphas=np.linspace(0.001,0.1,100))
l2_CV.fit(X_train_scaled, Y_train)
print('alpha = ', l2_CV.alpha_) 
print('coef = ', l2_CV.coef_)
```

```python
# Linear Regression - Training RMSE
y_pred_train_lr = lr.predict(X_train_scaled)
lr_train_MSE = mean_squared_error(Y_train, y_pred_train_lr)
lr_train_RMSE = sqrt(lr_train_MSE)
print('training MSE obtained in training dataset Linear Regression = ', lr_train_MSE)
# print('training RMSE obtained in training dataset Linear Regression = ', lr_train_RMSE)

# Linear Regression - testing RMSE
y_pred_test_lr = lr.predict(X_test_scaled)
test_MSE_lr = mean_squared_error(Y_test, y_pred_test_lr)
test_RMSE_lr = sqrt(test_MSE_lr)
print('testing MSE obtained in testing dataset Linear Regression = ', test_MSE_lr)
# print('testing RMSE obtained in testing dataset Linear Regression = ', test_RMSE_lr)

# Lasso (best alpha) - training RMSE
y_pred_train_l1 = l1_CV.predict(X_train_scaled)
train_MSE_l1 = mean_squared_error(Y_train, y_pred_train_l1)
train_RMSE_l1 = sqrt(train_MSE_l1)
print('\ntraining MSE obtained in training dataset Lasso Regression (best alpha) = ', train_MSE_l1)
# print('training RMSE obtained in training dataset Lasso Regression (best alpha) = ', train_RMSE_l1)

# Lasso (best alpha) - testing RMSE
y_pred_test_l1 = l1_CV.predict(X_test_scaled)
test_MSE_l1 = mean_squared_error(Y_test, y_pred_test_l1)
test_RMSE_l1 = sqrt(test_MSE_l1)
print('testing MSE obtained in testing dataset Lasso Regression (best alpha) = ', test_MSE_l1)
# print('testing RMSE obtained in testing dataset Lasso Regression (best alpha) = ', test_RMSE_l1)

# Ridge (best alpha) - training RMSE
y_pred_train_l2 = l2_CV.predict(X_train_scaled)
train_MSE_l2 = mean_squared_error(Y_train, y_pred_train_l2)
train_RMSE_l2 = sqrt(train_MSE_l2)
print('\ntraining MSE obtained in training dataset Ridge Regression (best alpha) = ', train_MSE_l2)
# print('training RMSE obtained in training dataset Ridge Regression (best alpha) = ', train_RMSE_l2)

# Ridge (best alpha) - testing RMSE
y_pred_test_l2 = l2_CV.predict(X_test_scaled)
test_MSE_l2 = mean_squared_error(Y_test, y_pred_test_l2)
test_RMSE_l2 = sqrt(test_MSE_l2)
print('testing MSE obtained in testing dataset Ridge Regression (best alpha) = ', test_MSE_l2)
# print('testing RMSE obtained in testing dataset Ridge Regression (best alpha) = ', test_RMSE_l2)
```

```python
# Print accuracy of models
print('\nAccuracy using Linear Regression= ', lr.score(X_test_scaled,Y_test))
print('\nAccuracy using Lasso  Regression and Cross validation= ', l1_CV.score(X_test_scaled,Y_test))
print('\nAccuracy using Ridge  Regression and Cross validation= ', l2_CV.score(X_test_scaled,Y_test))

# Got RidgeCV with maximum accuracy. Hence Ridge is chosen for final modelling
```
